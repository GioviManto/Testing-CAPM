# P1 #


```python


# Initialize an empty list to store data
data = []

# Iterate over each date
for date in p_1.index:
    # Extract market excess return for the date
    x = returns_2020.loc[date, 'Risk Premium']  # Market excess return
    
    # Extract portfolio excess return for the date and calculate mean
    y = p_1.loc[date].mean()  # Portfolio excess return
    
    # Append the values to the data list
    data.append({'Date': date, 'Market Excess Return': x, 'Portfolio Excess Return': y})

# Create DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Set Date column as index
df.set_index('Date', inplace=True)

# Display the DataFrame
df.head()
```

```python
import pandas as pd
import statsmodels.api as sm

# Create an empty list to store the coefficients
coefficients_data = []

# Iterate over each month
for month, group in df.groupby(pd.Grouper(freq='M')):
    # Extract the month's data
    monthly_data = group
    
    # Define independent and dependent variables
    X = monthly_data['Market Excess Return']  # Independent variable
    y = monthly_data['Portfolio Excess Return']  # Dependent variable

    monthly_portfolio_excess_return = monthly_data['Portfolio Excess Return'].values.mean()
    print(monthly_portfolio_excess_return)
    
    # Add constant to independent variables
    X = sm.add_constant(X)
    
    # Fit OLS model
    model = sm.OLS(y, X)
    results = model.fit()

    print(results.summary())

    # Extract the residuals from the regression results
    residuals = results.resid

    # Square the residuals to get the squared residuals
    squared_residuals = residuals ** 2

    # Calculate the variance of the squared residuals
    residual_variance = squared_residuals.var()

    print("Residual Variance:", residual_variance)
    
    # Store the coefficient for X1 in the list
    x1_coefficient = results.params['Market Excess Return']
    adj = x1_coefficient * 2/3 + 1/3
    # Extract the month part from the datetime object
    month_str = str(month.month).zfill(2)  # Convert month to string and ensure it's zero-padded
    
    coefficients_data.append({'Month': month_str, 'Portfolio Excess Return 1': monthly_portfolio_excess_return, 'Beta 1': x1_coefficient, 'Adjusted Beta 1': adj, "Residual Variance 1": residual_variance})

# Convert the list to a DataFrame
coefficients_df_port1 = pd.DataFrame(coefficients_data)

# Display the DataFrame with coefficients
coefficients_df_port1
```

```python
import statsmodels.api as sm
import pandas as pd

# Assuming new_df contains the required data

# Calculate adjusted beta squared
raw_beta_sqr = new_df['Beta'].values ** 2
adjusted_beta_sqr = 0.67 * raw_beta_sqr + 0.33 * 1.0

# Prepare the data
x = new_df['Adjusted Beta'].values
x_squared = adjusted_beta_sqr
residual_variance = new_df['Normalized Residual Variance'].values
y = new_df['Portfolio Excess Return'].values

# Prepare the independent variables matrix
X = sm.add_constant(pd.DataFrame({'Adjusted Beta': x, 'Adjusted Beta Squared': x_squared, 'Residual Variance': residual_variance}))

# Fit the nonlinear OLS model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary of results
print(results.summary())

# Retrieve specific values from the model summary
intercept = results.params[0]
gamma1 = results.params[1]  # Coefficient for adjusted beta
gamma2 = results.params[2]  # Coefficient for adjusted beta squared
gamma3 = results.params[3]  # Coefficient for residual variance
p_value_int = results.pvalues[0]
p_value_gamma1 = results.pvalues[1]  # P-value for gamma1
p_value_gamma2 = results.pvalues[2]  # P-value for gamma2
p_value_gamma3 = results.pvalues[3]  # P-value for gamma3
std_error_int = results.bse[0] 
std_error_gamma1 = results.bse[1]  # Standard error for gamma1
std_error_gamma2 = results.bse[2]  # Standard error for gamma2
std_error_gamma3 = results.bse[3]  # Standard error for gamma3
t_stat_int = results.tvalues[0]
t_stat_gamma1 = results.tvalues[1]  # T-statistic for gamma1
t_stat_gamma2 = results.tvalues[2]  # T-statistic for gamma2
t_stat_gamma3 = results.tvalues[3]  # T-statistic for gamma3

# Function to round the values
def round_to_six_decimals(value):
    return round(value, 6)

# Print the retrieved values rounded to the sixth decimal point
print("Intercept:", round_to_six_decimals(intercept))
print("Coefficient for adjusted beta (Î³1):", round_to_six_decimals(gamma1))
print("Coefficient for adjusted beta squared (Î³2):", round_to_six_decimals(gamma2))
print("Coefficient for residual variance (Î³3):", round_to_six_decimals(gamma3))
print("Std. error for intercept:", round_to_six_decimals(std_error_int))
print("Std. error for Î³1:", round_to_six_decimals(std_error_gamma1))
print("Std. error for Î³2:", round_to_six_decimals(std_error_gamma2))
print("Std. error for Î³3:", round_to_six_decimals(std_error_gamma3))
print("T-statistic for intercept:", round_to_six_decimals(t_stat_int))
print("T-statistic for Î³1:", round_to_six_decimals(t_stat_gamma1))
print("T-statistic for Î³2:", round_to_six_decimals(t_stat_gamma2))
print("T-statistic for Î³3:", round_to_six_decimals(t_stat_gamma3))
print("P-value for intercept:", round_to_six_decimals(p_value_int))
print("P-value for Î³1:", round_to_six_decimals(p_value_gamma1))
print("P-value for Î³2:", round_to_six_decimals(p_value_gamma2))
print("P-value for Î³3:", round_to_six_decimals(p_value_gamma3))
```
