R_it-R_ft= α_i+ β_i (R_mt-R_ft)+ ε_it         (6)
Where:
    R_it represents the rate of return on asset i at time t,
    R_ft denotes the risk-free rate at time t,
    R_mt signifies the rate of return on the market portfolio at time t,
    β_i  is the beta of stock i 
    ε_it represents the random disturbance term in the regression equation at time t.
    α_i  is the intercept term 
We can think of:
    (R_it-R_ft) as the excess return of stock i, and 
    (R_mt-R_ft ) as the average risk premium

```python
# Convert 'Date' column to datetime format
returns_df['Date'] = pd.to_datetime(returns_df['Date'])

# Extract year from 'Date' column
returns_df['Year'] = returns_df['Date'].dt.year

# Filter DataFrame to include only the data from 2018
returns_2018 = returns_df[returns_df['Year'] == 2018].copy()

# Drop the 'Year' column as it's no longer needed
returns_2018.drop(columns='Year', inplace=True)

```

```python
returns_2018.columns = returns_2018.columns.str.replace('2018-2023', '2018')
```

```python
import numpy as np

# Copy the 2018 data to avoid modifying the original DataFrame
returns_2018_excess = returns_2018.copy()

# Calculate excess returns for each stock by subtracting the risk-free rate
risk_free_rate = returns_2018_excess['BOT 3m'].values  # Convert Series to NumPy array
returns_2018_excess.iloc[:, 2:-2] -= risk_free_rate[:, np.newaxis]

# Display the first few rows of the new DataFrame
returns_2018_excess.head()
```

```python
import statsmodels.api as sm

# Extract excess returns of the asset and the market portfolio
x = returns_2018['FTSEMIB Index'] - returns_2018['BOT 3m']  # Market excess return
y = returns_2018['A2A 2018'] - returns_2018['BOT 3m']  # Asset excess return
    
 # Add a constant to the independent variable (market excess return)
x = sm.add_constant(x)
    
# Fit OLS model
model = sm.OLS(y, x)
results = model.fit()
    
print(results.summary())
```

```python
import statsmodels.api as sm

# Create an empty DataFrame to store the beta values
beta_df = pd.DataFrame(index=returns_2018.columns[1:-2], columns=['Beta'])

# Iterate over each asset
for asset in returns_2018_excess.columns[1:-2]:
    # Extract excess returns of the asset and the market portfolio
    x = returns_2018['FTSEMIB Index'] - returns_2018['BOT 3m']  # Market excess return
    y = returns_2018[asset] - returns_2018['BOT 3m']  # Asset excess return
    
    # Add a constant to the independent variable (market excess return)
    x = sm.add_constant(x)
    
    # Fit OLS model
    model = sm.OLS(y, x)
    results = model.fit()

    print(results.summary())
    
    # Extract the beta coefficient (slope)
    beta = results.params[0]  # Index 0 corresponds to the coefficient of the independent variable
    
    # Store the beta coefficient in the DataFrame
    beta_df.loc[asset, 'Beta'] = beta

# Display the DataFrame with beta values
beta_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a professional color palette
colors = sns.color_palette("Set2")

# Define colors based on beta values
colors = [colors[1] if (beta > 1 and index != 'FTSEMIB Index') else 
          colors[0] if (beta < 1 and index != 'FTSEMIB Index') else 
          'gray' if (beta == 1 or index == 'FTSEMIB Index') else 
          'red' for beta, index in zip(beta_df['Beta'], beta_df.index)]

# Plotting the beta values
plt.figure(figsize=(12, 8))
sns.barplot(x=beta_df.index, y=beta_df['Beta'], palette=colors)
plt.xlabel('Assets', fontsize=14)
plt.ylabel('Beta Values', fontsize=14)
plt.title('Beta Values for Each Asset in 2018', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)
plt.axhline(y=1, color='black', linestyle='-', linewidth=0.5)  # Add horizontal line at y=1
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines

# Create custom legend
legend_handles = [plt.Rectangle((0,0),1,1, color=sns.color_palette("Set2")[1]), 
                  plt.Rectangle((0,0),1,1, color=sns.color_palette("Set2")[0]), 
                  plt.Rectangle((0,0),1,1, color='grey')]

plt.legend(handles=legend_handles, 
           labels=['Beta > 1', 'Beta < 1', 'Beta = 1'], fontsize=12)

plt.tight_layout()
plt.show()
```
