```python
import pandas as pd

beta_df = pd.read_csv('/Users/gloriabagnato/Desktop/TESI/Data-Finance/Beta-2018.csv') 
```

```python
beta_df.rename(columns={'Unnamed: 0':'Stock'}, inplace=True)
```

```python
beta_df
```

```python
beta_df.drop(0, inplace=True)
```

```python
beta_df.sort_values(by="Beta")
```

```python
# Sort stocks based on beta values
sorted_df = beta_df.sort_values(by='Beta')

# Reset index after sorting
sorted_df = sorted_df.reset_index(drop=True)

# Divide stocks into four portfolios
portfolio_1 = sorted_df.head(10)
portfolio_2 = sorted_df.iloc[10:19]
portfolio_3 = sorted_df.iloc[19:28]
portfolio_4 = sorted_df.tail(10)
```

```python
print("Portfolio 1 (Lowest Beta):")
portfolio_1.to_csv("p1-1.csv")
portfolio_1
```

```python
print("Portfolio 2:")
portfolio_2.to_csv("p2-1.csv")
portfolio_2
```

```python
print("Portfolio 3:")
portfolio_3.to_csv("p3-1.csv")
portfolio_3
```

```python
print("Portfolio 4 (Highest Beta):")
portfolio_4.to_csv("p4-1.csv")
portfolio_4
```
