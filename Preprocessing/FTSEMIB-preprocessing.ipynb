```python
import pandas as pd

ftsemib_df = pd.read_excel('/Users/gloriabagnato/Desktop/TESI/Data-Finance/FTSEMIB/FTSE MIB 2018-2023.xlsx')

ftsemib_df
```

```python
ftsemib_df.rename(columns={'FTSEMIB Index - Ultimo prz':'FTSEMIB Index - Last price'}, inplace=True)
```

```python
ftsemib_df
```

```python
ftsemib_df.to_csv('ftsemib_filtered.csv', index=False)
```

```python
# Convert 'Date' column to datetime if it's not already in datetime format
ftsemib_df['Date'] = pd.to_datetime(ftsemib_df['Date'])

# Sort DataFrame by date in ascending order
ftsemib_df = ftsemib_df.sort_values(by='Date')

# Calculate returns
ftsemib_df['Returns'] = ftsemib_df['FTSEMIB Index - Last price'].pct_change()

# Drop the first row since it doesn't have a previous value for calculating returns
ftsemib_df = ftsemib_df.dropna()

# Display the DataFrame with returns
ftsemib_df
```

```python
# Format returns as percentage with two decimal places
ftsemib_df['Returns'] = (ftsemib_df['Returns'] * 100).map('{:.2f}%'.format)

ftsemib_df
```

```python
# Calculate the standard deviation
std = ftsemib_df['FTSEMIB Index - Last price'].std()

print("Standard Deviation:", "{:.2f}".format(std))  

# Calculate the mean
mean = ftsemib_df['FTSEMIB Index - Last price'].mean()

print("Mean:", "{:.2f}".format(mean))  

# Calculate the absolute mean
mean_abs = abs(mean)

print("Absolute mean:", "{:.2f}".format(mean_abs))  

# Calculate the coefficient of variation
cv = std / mean_abs * 100

# Print the coefficient of variation with percentage format
print("Coefficient of Variation:", "{:.2f}%".format(cv)) 
```

```python
from docx import Document

# Create a new Document
doc = Document()

# Add a table to the document
table = doc.add_table(rows=1, cols=2)

# Add header row
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Statistic'
hdr_cells[1].text = 'Value'

# Data
data = {
    'Count': 312.000000,
    'Mean': 23268.517019,
    'Standard Deviation': 3154.023593,
    'Minimum': 15731.850000,
    '25th Percentile': 21017.372500,
    'Median': 22949.975000,
    '75th Percentile': 25711.525000,
    'Maximum': 30403.900000
}

# Add data rows to the table
for key, value in data.items():
    row_cells = table.add_row().cells
    row_cells[0].text = key
    row_cells[1].text = "{:.2f}".format(value)

# Save the document
doc.save('ftsemib_statistics.docx')
```
