```python
import pandas as pd

# Assuming df is your DataFrame containing the provided data

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Filter DataFrame for the specified date interval
filtered_df = df[(df['Date'] >= '2018-01-01') & (df['Date'] <= '2023-12-31')]

# Select the column 'GTITL3MO @BGN Corp (R1)' and rename it to 'BOT 3m'
filtered_df = filtered_df[['Date', 'GTITL3MO @BGN Corp  (R1)']].rename(columns={'GTITL3MO @BGN Corp  (R1)': 'BOT 3m'})

# Reset index
filtered_df.reset_index(drop=True, inplace=True)

filtered_df
```
