```python
import pandas as pd

ft_daily = pd.read_excel('/Users/gloriabagnato/Desktop/TESI/Data-Finance/FTSEMIB/FTSEMIB Daily 2018-2023.xlsx')

ft_daily
```

```python
# Convert the 'Date' column to datetime format
ft_daily['Date'] = pd.to_datetime(ft_daily['Date'])

# Sort the DataFrame by date in ascending order
ft_daily.sort_values('Date', inplace=True)

# Calculate the daily returns
ft_daily['Returns'] = ft_daily['Ultimo prz'].pct_change()

# Drop the first row since it doesn't have a previous value to calculate the return
ft_daily.dropna(subset=['Returns'], inplace=True)

# Now, you can calculate the variance of the returns
returns_variance = ft_daily['Returns'].var()

print("Variance of returns:", returns_variance)
```
