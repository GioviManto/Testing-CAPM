```python
returns_df.drop(columns='Unnamed: 0', inplace = True)
```

```python
# Convert 'Date' column to datetime format
returns_df['Date'] = pd.to_datetime(returns_df['Date'])

# Extract year from 'Date' column
returns_df['Year'] = returns_df['Date'].dt.year

# Filter DataFrame to include only the data from 2021
returns_2021 = returns_df[returns_df['Year'] == 2021].copy()

# Drop the 'Year' column as it's no longer needed
returns_2021.drop(columns='Year', inplace=True)

```

```python
returns_2021.columns = returns_2021.columns.str.replace('2018-2023', '2021')
```

```python
import numpy as np

# Copy the 2021 data to avoid modifying the original DataFrame
returns_2021_excess = returns_2021.copy()

# Calculate excess returns for each stock by subtracting the risk-free rate
risk_free_rate = returns_2021_excess['BOT 3m'].values  # Convert Series to NumPy array
returns_2021_excess.iloc[:, 2:-2] -= risk_free_rate[:, np.newaxis]

# Display the first few rows of the new DataFrame
returns_2021_excess.head()
```

```python
import statsmodels.api as sm

# Create an empty DataFrame to store the beta values
beta_df = pd.DataFrame(index=returns_2021.columns[1:-2], columns=['Beta'])

# Iterate over each asset
for asset in returns_2021_excess.columns[1:-2]:
    # Extract excess returns of the asset and the market portfolio
    x = returns_2021['FTSEMIB Index'] - returns_2021['BOT 3m']  # Market excess return
    y = returns_2021[asset] - returns_2021['BOT 3m']  # Asset excess return
    
    # Add a constant to the independent variable (market excess return)
    x = sm.add_constant(x)
    
    # Fit OLS model
    model = sm.OLS(y, x)
    results = model.fit()

    print(results.summary())
    
    # Extract the beta coefficient (slope)
    beta = results.params[0]  # Index 0 corresponds to the coefficient of the independent variable
    
    # Store the beta coefficient in the DataFrame
    beta_df.loc[asset, 'Beta'] = beta

# Display the DataFrame with beta values
beta_df
```

```python
beta_df.drop(index='FTSEMIB Index', inplace=True)
```

# P1 #

```python
import pandas as pd

# Initialize an empty list to store data
data = []

# Iterate over each date
for date in p_1.index:
    # Extract market excess return for the date
    x = returns_2022.loc[date, 'Risk Premium']  # Market excess return
    
    # Extract portfolio excess return for the date and calculate mean
    y = p_1.loc[date].mean()  # Portfolio excess return
    
    # Append the values to the data list
    data.append({'Date': date, 'Market Excess Return': x, 'Portfolio Excess Return': y})

# Create DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Set Date column as index
df.set_index('Date', inplace=True)

# Display the DataFrame
df.head()
```

```python
import pandas as pd
import statsmodels.api as sm

# Create an empty list to store the coefficients
coefficients_data = []

# Iterate over each month
for month, group in df.groupby(pd.Grouper(freq='M')):
    # Extract the month's data
    monthly_data = group
    
    # Define independent and dependent variables
    X = monthly_data['Market Excess Return']  # Independent variable
    y = monthly_data['Portfolio Excess Return']  # Dependent variable

    monthly_portfolio_excess_return = monthly_data['Portfolio Excess Return'].values.mean()
    print(monthly_portfolio_excess_return)
    
    # Add constant to independent variables
    X = sm.add_constant(X)
    
    # Fit OLS model
    model = sm.OLS(y, X)
    results = model.fit()

    print(results.summary())
    
    # Store the coefficient for X1 in the list
    x1_coefficient = results.params['Market Excess Return']
    adj = x1_coefficient * 2/3 + 1/3
    # Extract the month part from the datetime object
    month_str = str(month.month).zfill(2)  # Convert month to string and ensure it's zero-padded
    
    coefficients_data.append({'Month': month_str, 'Portfolio Excess Return': monthly_portfolio_excess_return, 'Beta': x1_coefficient, 'Adjusted Beta': adj})

# Convert the list to a DataFrame
coefficients_df = pd.DataFrame(coefficients_data)

# Display the DataFrame with coefficients
coefficients_df
```

```python
coefficients_df.to_csv('coefficients_df_p1_4.csv') 
```

```python
import statsmodels.api as sm
import pandas as pd

x = df['Market Excess Return'].values # Market excess return

y = df['Portfolio Excess Return'].values # Portfolio excess return
   
    
# Add a constant to the independent variable (market excess return)
x = sm.add_constant(x)

print(x.shape)
    
# Fit OLS model
model = sm.OLS(y, x)
results = model.fit()
    
print(results.summary())
    
# Retrieve specific values from the model summary
beta_p = results.params[1]  # Coefficient
p_value = results.pvalues[1]  # P-value
std_error = results.bse[1]  # Standard error
adj_r2 = results.rsquared_adj  # Adjusted R-squared
f_stat = results.fvalue  # F-statistic
t_stat_x1 = results.tvalues[1]

# Print the retrieved values
print("β_p (Coefficient):", beta_p)
print("P-value:", p_value)
print("Std. error:", std_error)
print("Adj. R-squared:", adj_r2)
print("F-statistic:", f_stat)
print("T-statistic for X1:", t_stat_x1)
#
```
